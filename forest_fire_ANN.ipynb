```python
df = pd.read_csv('forestfires.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lab_enco = LabelEncoder()
df['month']=lab_enco.fit_transform(df['month'])
df['day']=lab_enco.fit_transform(df['day'])
df['size_category']= lab_enco.fit_transform(df['size_category']) 
                                 
```

```python
x = df.iloc[:,:11]
y = df.iloc[:,-1]
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(20,input_dim=11,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
result = model.fit(x,y,validation_split=0.2,epochs=25,batch_size = 15)
```

```python
model.evaluate(x,y)
```

```python
predict = model.predict_on_batch(x)
comparison= pd.DataFrame({'actual_category':y})
comparison['predicted_category'] = np.round(predict)
comparison
```

```python
np.mean(comparison['actual_category']==comparison['predicted_category'])
```
